```python
import pandas as pd

df = pd.read_csv('c:/Users/Chakkaphan/Downloads/product_backlog_2026-02-01 (1).csv')

df = df.drop(columns=["Priority","Assignee"])


output_path = 'week01.txt'

with open(output_path, 'w', encoding='utf-8') as f:
    for i, row in df.iterrows():
        if pd.isna(row['Status']) or row['Status'].strip() == '' or row['Status'] == "productbacklog":
            continue
        f.write(f"{i+1}. {row['Title']} - {row['Description']}  ||   {row['Status']}\n")

print("เขียนไฟล์ product_backlog.txt เรียบร้อยแล้ว")
```
